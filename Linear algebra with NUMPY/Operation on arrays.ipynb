```python
import numpy as np

arr = np.array([1, 2, 3, 4])
```

```python
np.sum(arr)  # 1 + 2 + 3 + 4 = 10
```

```python
arr.sum()
```

```python
np.abs(-2)  #|2| модуль
```

```python
np.prod(arr)  # 1 * 2 * 3 * 4 = 24
```

```python
np.sqrt(arr)  # квадратный корень
```

```python
ar = np.array([4.5, 1.2])
np.modf(ar)  # Дробные и целые части массива в виде отдельных массивов
```

```python
# Массив логических (булевых) значений, показывающий, какие из элементов исходного массива являются NaN (не числами)
np.isnan(arr)
```
