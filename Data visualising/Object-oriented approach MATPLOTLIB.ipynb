```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('Restaurant.csv')
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.hist(df['total_bill'])
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.hist(df['total_bill'], bins = 30, color = 'green')
axes.set_title('The total amount of order')
axes.set_ylabel('Quantity of orders')
axes.set_xlabel('Amount of order, USD')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 2, 1])
axes.scatter(x = df['total_bill'], y = df['tip'], marker = '+', color = 'black')
axes.set_title('The total amount of order Vs Amount of tip')
axes.set_ylabel('Amount of bill')
axes.set_xlabel('Amount of tip')
```

```python
tips = df.pivot_table(values = ['total_bill', 'tip'],
                     index = 'day',
                     aggfunc = 'mean')
tips.head()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.set_ylim(0, 30)

axes.bar(x = tips.index, height = tips['total_bill'],
        width = 0.4, align = 'edge', label = 'Average bill')
axes.bar(x = tips.index, height = tips['tip'],
        width = -0.4, align = 'edge', label = 'Average tip')

axes.set_title('Average bill Vs Average tip by day')
axes.set_ylabel('Average')
axes.set_xlabel('Day of week')

axes.legend(loc = 1)
```

```python
fig, axes = plt.subplots(2, 3, figsize = (20, 10))
```
