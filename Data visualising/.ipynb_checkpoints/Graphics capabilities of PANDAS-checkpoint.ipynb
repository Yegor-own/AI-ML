```python
import pandas as pd

df = pd.read_csv('Restaurant.csv')
df.head()
```

```python
df.plot()
```

```python
df['total_bill'].plot(kind = 'hist')
```

```python
df['total_bill'].plot(kind = 'hist', grid = True, title = 'Sum of bill')
```

```python
df['day'].value_counts().plot(kind = 'bar',
                             grid = True,
                             colormap = 'coolwarm',
                             title = 'Quantity of clients by day')
```

```python
df[['total_bill', 'tip']].plot(kind = 'hist',
                               grid = True,
                               subplots = True,
                               title = ['Sum of bill', 'Sum of tips'],
#                                legend = False
                              )
```

```python
df.groupby(by = ['sex', 'vegan'])['tip'].mean().plot(kind = 'bar', grid= True)
```
