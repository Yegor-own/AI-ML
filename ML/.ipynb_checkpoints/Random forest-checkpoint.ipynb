# Random Forest

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

### Загружаем данные

Будем использовать встроенный датасет о диабете

```python
# Загружаем датасет
diabetes = load_diabetes()
```

**Описание колонок по-русски:**

    - age - возраст
    - sex - пол
    - bmi - индекс массы тела
    - bp - средняя кровяное давление
    - s1-s6 - показатели крови
    - target - ЦП, мера прогресса заболевания через год после сбора данных 

```python
# Кладем его в табличку
diabetes_df = pd.DataFrame(data=diabetes['data'], index=range(len(diabetes['data'])), columns=diabetes['feature_names'])
diabetes_df['target'] = diabetes['target']
```

```python
# Смотрим, что получилось
diabetes_df.head()
```

Подумайте, как можно обработать данные


```python
diabetes_df['sex'].value_counts().index[0]
diabetes_df['sex'] = diabetes_df['sex'].map({diabetes_df['sex'].value_counts().index[1]:0, diabetes_df['sex'].value_counts().index[0]:1})
```

### Деление данных на train/test

```python
target = diabetes_df.target
data = diabetes_df.drop('target', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=322, test_size=0.1)
```

### Обучение модели

```python
# Создаем модель
rf = RandomForestRegressor(n_estimators=500, n_jobs=-1, random_state=322)
# Обучаем
rf.fit(X_train, y_train)
```

### Оценка качества модели

```python
# Качество модели на той выборке, на которой она обучалась
print('Train Accuracy:', mean_squared_error(y_train, rf.predict(X_train)))
# Качество на тестовой выборке
print('Validation Accuracy:', mean_squared_error(y_test, rf.predict(X_test)))
```

### Отбор признаков

Одна из задач, которую RF умеет решать и очень неплохо - это отбор признаков

```python
# Упорядычиваем наши фичи по значениям весов, от самой полезной к самой бесполезной
df_importances= sorted(list(zip(data.columns, rf.feature_importances_.ravel())), key=lambda tpl: tpl[1], reverse=True)

# Создаем табличку, в которой будет показан признак и его вес
df_importances = pd.DataFrame(df_importances, columns=['feature', 'importance']) 

# Нумируем колонки, чтобы не путать их
df_importances = df_importances.set_index('feature')

# Создаем график, чтобы было нагляднее
df_importances.plot(kind='bar', figsize=(15, 3))

# Рисуем график
plt.show()

# Выводим табличку
display(df_importances.head(10))
```

### Оставляем только важные признаки

```python
imp = ['bmi', 's5', 'bp', 's6', 's2']

# Разобьем датасет
target = diabetes_df.target
data = diabetes_df[imp]

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=322, test_size=0.25)

# Создаем модель
rf_2 = RandomForestRegressor(n_estimators=500, n_jobs=-1, random_state=322)
# Обучаем
rf_2.fit(X_train, y_train)

# Качество модели на той выборке, на которой она обучалась
print('Train Accuracy:', mean_squared_error(y_train, rf_2.predict(X_train)))
# Качество на тестовой выборке
print('Validation Accuracy:', mean_squared_error(y_test, rf_2.predict(X_test)))
```

### Посмотрим на результат при использовании модели линейной регресии

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
print('Train Accuracy:', mean_squared_error(y_train, lin_reg.predict(X_train)))
# Теперь оцените качество на тестовой выборке
print('Validation Accuracy:', mean_squared_error(y_test, y_pred))
```
