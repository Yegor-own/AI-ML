```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

Проведите анализ данных, оставив те столбцы, которые считаете ключевыми для обучения модели и преобразуйте все категориальные признаки в числовое значение

```python
data['y'] = data.y.astype('category').cat.codes
```

```python
data.info()
```

```python
data = data.drop('marital', axis=1)
data = data.drop('contact', axis=1)
data = data.drop('day', axis=1)
data = data.drop('month', axis=1)
data = data.drop('campaign', axis=1)
data = data.drop('pdays', axis=1)
data = data.drop('previous', axis=1)
```

```python
cat_features = ['job', 'education', 'housing', 'loan', 'poutcome']

for i in cat_features:
    data[i] = data[i].astype('category').cat.codes
```

```python
data.head() #проверьте свой результат, что остались только необходимые столбцы
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('y', axis=1), data['y'], test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
LogReg = LogisticRegression(random_state=0).fit(X_train, y_train)
```

Сделайте своё предсказание 

```python
y_pred_train = LogReg.predict(X_train)
y_predict_proba_train = LogReg.predict_proba(X_train)[:,1]
```

```python
LogReg.score(X_train, y_train)
```

Проверьте результаты обученной модели, используя метрику ROC-AUC

```python
from sklearn.metrics import roc_auc_score
print('ROC-AUC: ', roc_auc_score(y_train, y_predict_proba_train))
```
